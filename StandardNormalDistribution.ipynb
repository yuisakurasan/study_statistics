# 標準正規分布

確率密度関数

$$
f(x)=\dfrac{1}{\sqrt{2π}×xの標準偏差}e^{-\dfrac{1}{2}(\dfrac{x-xの平均}{xの標準偏差})~2}=\dfrac{1}{\sqrt{2π}×1}e^{-\dfrac{1}{2}(\dfrac{x-0}{1})^2}=\dfrac{1}{\sqrt{2π}}e^{-\dfrac{1}{2}x^2}
$$

xは，標準正規分布に従う

正規分布の時のデータを基準化した後，そのテスト結果は標準正規分布に従う

確率密度分布のグラフと横軸で囲まれた部分の面積は1

標準正規分布の面積と，割合・確率は同一視できる．

面積＝割合＝確率

標準正規分布は、世界中のありとあらゆるデータを**「平均0からの距離（$Z$値）」という共通言語に翻訳し、「そのデータが起きる確率は何％か？」**を判定するために使われます。

```python
import scipy.stats as stats
```

```python
# 1. データの条件
mean = 170
std = 5
target_value = 180

# 2. Z値を求める（標準化）
# 式: (データ - 平均) / 標準偏差
z = (target_value - mean) / std
print(f"Z値（標準偏差いくつ分か）: {z}") 
# 結果は 2.0 になるはず（平均より+2σの位置）

# 3. 標準正規分布を使って確率を求める
lower_prob = stats.norm.cdf(z, loc=0, scale=1) 

# 上位何%か知りたいので、全体(1.0)から引く
upper_prob = 1.0 - lower_prob

print(f"身長{target_value}cmは、下から {lower_prob:.2%} の位置です。")
print(f"つまり、上位 {upper_prob:.2%} に入ります。")
```
